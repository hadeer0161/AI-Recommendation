```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# --- 1. DATASET SETUP (Books & Genres) ---
# Mock dataset representing items available for recommendation
data = {
    'Item_ID': [1, 2, 3, 4, 5, 6, 7, 8],
    'Title': [
        'The Hobbit', 'Neuromancer', 'The Da Vinci Code',
        'Dune', 'Sherlock Holmes', 'Dracula',
        'Foundation', 'The Silent Patient'
    ],
    'Genres': [
        'Fantasy Adventure Myth',
        'Sci-Fi Cyberpunk Technology',
        'Mystery Thriller Suspense',
        'Sci-Fi Space-Opera Fantasy',
        'Mystery Detective Crime',
        'Horror Vampire Gothic',
        'Sci-Fi Space-Opera Empire',
        'Mystery Psychological Thriller'
    ]
}

df = pd.DataFrame(data)

print("--- System Initialized ---")
print(f"Loaded {len(df)} library items across Sci-Fi, Fantasy, Mystery, and Horror.\n")

# --- 2. TAKE USER INPUT (Preferences) ---
print("What are your reading interests? (e.g., 'Sci-Fi Space-Opera', 'Mystery Thriller Detective')")
user_input = input("Enter keywords/genres: ")

if not user_input.strip():
    # Fallback default if input is empty
    user_input = "Sci-Fi Space-Opera"
    print(f"No input detected. Using default preferences: '{user_input}'")

# Append user preferences as a temporary row to the dataframe for vectorization
user_row_index = len(df)
df.loc[user_row_index] = [0, 'User_Profile', user_input]

# --- 3. PATTERN MATCHING & SIMILARITY LOGIC ---
# Convert the text descriptions/genres into a matrix of token counts
vectorizer = CountVectorizer()
count_matrix = vectorizer.fit_transform(df['Genres'])

# Calculate the Cosine Similarity matrix between all items (including the user profile)
similarity_matrix = cosine_similarity(count_matrix)

# Extract the similarity scores specifically matching the user profile row
user_scores = similarity_matrix[user_row_index]

# Map scores to items, excluding the user profile itself
# Create tuples of (item_index, similarity_score)
score_pairs = list(enumerate(user_scores[:-1]))

# Sort items based on similarity scores in descending order
sorted_recommendations = sorted(score_pairs, key=lambda x: x[1], reverse=True)

# --- 4. DISPLAY RECOMMENDED ITEMS ---
print("\n=============================================")
print(f" TOP RECOMMENDATIONS FOR YOUR PROFILE ")
print("=============================================")

recommendation_count = 0
for index, score in sorted_recommendations:
    # Only recommend items that have some degree of match/similarity (score > 0)
    if score > 0:
        recommendation_count += 1
        title = df.iloc[index]['Title']
        genres = df.iloc[index]['Genres']
        # Convert similarity score to a clean percentage string
        match_percentage = f"{score * 100:.1f}%"

        print(f"{recommendation_count}. {title:<20} | Match: {match_percentage:<6} | Tagged: {genres}")

    # Cap results at the top 3 recommendations
    if recommendation_count >= 3:
        break

if recommendation_count == 0:
    print("No direct matches found for your exact keywords. Try broader genres like 'Sci-Fi' or 'Mystery'!")
print("=============================================")
```
